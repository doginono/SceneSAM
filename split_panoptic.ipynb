```python
def convert_image_name(image_name):
    # Split the image name at the underscore
    prefix, number = image_name.split('_')
    
    # Pad the number with one additional zero
    padded_number = f'{int(number):06d}'  # Ensures the number has 5 digits
    
    # Create the new image names
    frame_name = f'frame{padded_number}'
    depth_name = f'depth{padded_number}'
    
    return frame_name, depth_name

# Example usage
original_name = '0_0001'
frame_name, depth_name = convert_image_name(original_name)

print('Frame:', frame_name)
print('Depth:', depth_name)
```

```python
frames, depths = [], []
for d in data:
    frame_name, depth_name = convert_image_name(d)
    frames.append(frame_name+'.jpg')
    depths.append(depth_name+'.png')

print(frames[-5:])
print(depths[-5:])
```

```python
# Paths to the directories
source_folder = '/home/rozenberszki/project/wsnsl/Datasets/Replica/room0_panoptic/results'
destination_folder = '/home/rozenberszki/project/wsnsl/Datasets/Replica/room0_panoptic/test_results_org_pan'

# Ensure the destination folder exists
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# List of image names in the subset
subset_list = depths  # Add your image names here
```

```python
# Loop through all files in the source folder
for file_name in os.listdir(source_folder):
    if file_name not in frames and file_name not in depths:
        # Full path of the file
        source_file_path = os.path.join(source_folder, file_name)
        destination_file_path = os.path.join(destination_folder, file_name)
        
        # Move the file
        shutil.move(source_file_path, destination_file_path)
        print(f'Moved: {file_name}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random images
image1 = np.random.random((100, 100))
image2 = np.random.random((100, 100))
image3 = np.random.random((100, 100))

# Create a subplot with 1 row and 3 columns
fig, axes = plt.subplots(1, 3)
fig.subplots_adjust(left=0, right=1, top=1, bottom=0)
# Plot the images on the subplots
axes[0].imshow(image1, cmap='gray')
axes[0].axis('off')

axes[1].imshow(image2, cmap='gray')
axes[1].axis('off')

axes[2].imshow(image3, cmap='gray')
axes[2].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
fig.savefig('test.png', bbox_inches='tight',pad_inches=0)
```
